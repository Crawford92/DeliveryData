```python
import pandas as pd
dashDF = pd.read_csv('entries.csv')
dashdaysDF = pd.read_csv('days.csv')
dashweeksDF = pd.read_csv('weeks.csv')
```

```python
dashDF.head()
```

```python
dashDF["Restaurant Name"].nunique()
```

```python
dashDF["Pay"] = dashDF["Pay"].str.replace('$', '')
dashDF["Tip"] = dashDF["Tip"].str.replace('$', '')
dashDF["Total"] = dashDF["Total"].str.replace('$', '')
```

```python
dashDF.head()
```

```python
dashDF["Pay"].unique()
```

```python
dashDF["Pay"] = dashDF["Pay"].astype('float')
```

```python
dashDF["Tip"] = dashDF["Tip"].astype('float')
dashDF["Total"] = dashDF["Total"].astype('float')
```

```python
dashDFbyStoreAVG = dashDF.groupby(["Restaurant Name"]).agg({"Pay": "mean", "Tip": "mean"})
#Pay": "sum", "Tip": "sum", 
```

```python
dashDFbyStoreAVG = dashDFbyStoreAVG.round({'Pay': 2, 'Tip': 2})
```

```python
dashDFbyStoreAVG = dashDFbyStoreAVG.sort_values("Tip", ascending=False)
dashDFbyStoreAVG.head(20)
```
